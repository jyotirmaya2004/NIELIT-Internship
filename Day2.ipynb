## Type casting
#### Python supports conversion of one data type to another

#### Type conversion :

1.   Implicit conversion
2.   Explicit conversion





### Implicit conversion

```python
a=5
b=6.5
print(a+b)
```

```python
a=5
b=2.5
c=a+b # int is automatically  to float
print(c)
print(type(c))
```

### Explicit conversion




1.   int() : convert values into integer
2.   float() : converts value into float
3.   complex() : converts value to complex number
4.   str() : converts value to string
5.   list() : converts value to list
6.   tuple() : converts value to tuple
7.   set() : converts value to set
8.   dict() : converts value to dictionary
9.   bool() : converts value to boolean
10.  chr () : converts ASCII/unicode in to character
11.  ord() : converts value to unicode/ASCII





## int() : convert values into integer

```python
a= int(3.5)
print(a)
print(type(a))
```

## float() : convert values into float

```python
a= float(3)
print(a)
print(type(a))
```

```python
a= complex(3,5)
print(a)
print(type(a))
```

```python
a= str(34)
print(a)
print(type(a))
```

```python
a= list((3,6,5,4,5))
print(a)
print(type(a))
```

```python
a= tuple([3,6,5,4,5])
print(a)
print(type(a))
```

```python
a= set([3,6,5,4,5])
print(a)
print(type(a))
```

```python
a= dict([(1,"One"),(2,"Two"),(3,"Three")])
print(a)
print(type(a))
```

```python
a= bool(0)
print(a)
print(type(a))
```

```python
a= chr(65)
print(a)
print(type(a))
```

```python
a= ord('A')
print(a)
print(type(a))
```

## Operators in python

1. Arithmatic(+, -, *, /, %)
2. Relational(Comparision)(==, !=, >, <, >=, <=)
3. Assignment(=, +=, -=, *=, %=, /=)
4. Logical (and, or, not)
5. Bitwise (&,|,^,-,<<,>>)
6. Membership(in, not in)
7. Identity (is, is not)



## Precedence of operator

| Sl. No. | Precedence Level | Operators                        | Description                             | Example          |
| ------- | ---------------- | -------------------------------- | --------------------------------------- | ---------------- |
| 1       | Highest          | `()`                             | Parentheses                             | `(2+3)*4`        |
| 2       | High             | `**`                             | Exponentiation (power)                  | `2**3 = 8`       |
| 3       | High             | `+x`, `-x`, `~x`                 | Unary plus, minus, bitwise NOT          | `-5`, `~2`       |
| 4       | High             | `*`, `/`, `//`, `%`              | Multiply, divide, floor divide, modulus | `10//3`          |
| 5       | Medium           | `+`, `-`                         | Addition, subtraction                   | `5+2`            |
| 6       | Medium           | `<<`, `>>`                       | Bitwise shift                           | `4<<1`           |
| 7       | Medium           | `&`                              | Bitwise AND                             | `5 & 3`          |
| 8       | Medium           | `^`                              | Bitwise XOR                             | `5 ^ 3`          |
| 9       | Medium           | `\|`                             | Bitwise OR                              | `5 \| 3`         |
| 10      | Low              | `==`, `!=`, `>`, `<`, `>=`, `<=` | Comparison operators                    | `a > b`          |
| 11      | Low              | `not`                            | Logical NOT                             | `not True`       |
| 12      | Lower            | `and`                            | Logical AND                             | `True and False` |
| 13      | Lowest           | `or`                             | Logical OR                              | `True or False`  |
| 14      | Very Low         | `=`, `+=`, `-=`, `*=`, etc.      | Assignment operators                    | `a += 5`         |
