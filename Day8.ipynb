### Scope in python

##### Scope refers to the coding region from which a particular python object is accessible. Hence one can't access any particular object from anywhere from the code, the accessing has to be allowed by the scope of the object.

##### **Types of scope**

| Scope Type          | Meaning                                             | Accessible Where          | Example                     |
| ------------------- | --------------------------------------------------- | ------------------------- | --------------------------- |
| **Local Scope**     | Variable declared inside a function                 | Only inside that function | `x = 10` inside `def fun()` |
| **Enclosing Scope** | Variable from outer function used in inner function | Inside nested function    | Outer function variable     |
| **Global Scope**    | Variable declared outside all functions             | Whole program             | `x = 100` outside function  |
| **Built-in Scope**  | Predefined names provided by Python                 | Available everywhere      | `print()`, `len()`, `sum()` |


```python
x="global"
def outer():
  x="enclosing"
  def inner():
    x="local"
    print(x)
  inner()
  print(x)
outer()
print(x)
```

```python
x = 100   # Global

def outer():
    y = 50   # Enclosing

    def inner():
        z = 20   # Local
        print("Local : ",z)   # Local
        print("Enclosing : ",y)   # Enclosing
        print("Global : ",x)   # Global
        print("Built-in : ",len("abc"))   # Built-in

    inner()

outer()
```

######11. split(), join()

| Method    | Purpose                           | Syntax                    | Example                   | Output            |
| --------- | --------------------------------- | ------------------------- | ------------------------- | ----------------- |
| `split()` | Splits a string into a list       | `string.split(separator)` | `"a b c".split()`         | `['a', 'b', 'c']` |
| `join()`  | Joins list elements into a string | `"sep".join(list)`        | `" ".join(['a','b','c'])` | `"a b c"`         |


```python
text = "Python is easy"
words = text.split()
print(words)
```

```python
text = "apple,mango,banana"
fruits = text.split(",")
print(fruits)
```

```python
words = ['Python', 'is', 'easy']
text = " ".join(words)
print(text)
```
