## Lambda function

In Python, a **lambda function** (also known as an anonymous function) is a small, single-expression function that can be defined without a name. It's often used for short, throwaway functions that are not intended to be reused elsewhere.

**Key characteristics:**
-   Defined using the `lambda` keyword.
-   Can take any number of arguments, but can only have one expression.
-   The expression is implicitly returned.
-   Cannot contain statements or annotations other than the expression.

````python
add = lambda x, y: x + y
print(f"Sum of 5 and 3: {add(5, 3)}")
````

### Regular Functions vs. Lambda Functions

| Feature            | Regular Functions (defined with `def`)                      | Lambda Functions (defined with `lambda`)                |
|:-------------------|:----------------------------------------------------------|:--------------------------------------------------------|
| **Keyword**        | `def`                                                     | `lambda`                                                |
| **Name**           | Have a name                                               | Anonymous (no name)                                     |
| **Statements**     | Can contain multiple expressions and statements (e.g., `if`, `for`, `try-except`) | Can only contain a single expression                    |
| **Return Value**   | Use `return` keyword to explicitly return a value         | Implicitly returns the result of the single expression  |
| **Complexity**     | Suitable for complex logic                                | Suitable for simple, single-expression operations       |
| **Reusability**    | Designed for reuse                                        | Often used for throwaway, one-time use                  |
| **Docstrings/Annotations** | Can have docstrings and type annotations                  | Cannot have docstrings or type annotations              |
| **Scope**          | Can be defined globally or locally                        | Typically defined inline where needed (e.g., as arguments to higher-order functions) |
| **Use Cases**      | General-purpose functions, complex algorithms             | Short callbacks, arguments for `map()`, `filter()`, `sorted()`, `key` arguments |


#####A higher-order function works with other functions (either receives them or returns them).

### Lambda with Higher-Order Functions

| Higher-Order Function | Lambda Function Used   | Purpose                                     | Original Input                 | Output                                   |
|:----------------------|:-----------------------|:--------------------------------------------|:-------------------------------|:-----------------------------------------|
| `map()`               | `lambda x: x**2`       | Square each number in a list                | `[1, 2, 3, 4, 5]`              | `[1, 4, 9, 16, 25]`                      |
| `filter()`            | `lambda x: x % 2 == 0` | Select even numbers from a list             | `[1, ..., 10]`                 | `[2, 4, 6, 8, 10]`                       |
| `sorted()`            | `lambda student: student['age']` | Sort list of dictionaries by 'age'        | `[{'name': 'Alice', ...}, ...]`| `[{'name': 'Bob', 'age': 20, ...}, {'name': 'Charlie', 'age': 25, ...}, {'name': 'Alice', 'age': 30, ...}]` |
| `sorted()`            | `lambda student: student['grade']` | Sort list of dictionaries by 'grade'      | `[{'name': 'Alice', ...}, ...]`| `[{'name': 'Bob', 'age': 20, 'grade': 'A'}, {'name': 'Alice', 'age': 30, 'grade': 'B'}, {'name': 'Charlie', 'age': 25, 'grade': 'C'}]` |


#### map()

###### map() applies a function to each element of an iterable.

````python
nums=[2,3,4,5,6,7,8]
squares=list(map(lambda x:x**2,nums))
print(squares)
print(map(lambda x:x**2,nums))
````

#### filter()

######filter() keeps only elements where function returns True

````python
nums=[2,3,4,5,6,7,8]
evens=list(filter(lambda x:x%2==0,nums))
print(evens)
print(filter(lambda x:x%2==0,nums))
````

#### sorted()

######sorted() arranges elements in order (default: ascending)

````python
students=[('Alice',85),('Bob',92),('tommy',35),('Michel',99)]
ranked=sorted(students,key=lambda x:x[1])
print(ranked)
print(sorted(students,key=lambda x:x[1],reverse=True))
````

````python
str1=['a','ab','bcd','ad']
ranked=sorted(str1,key=len)
print(ranked)
print(sorted(str1,key=len,reverse=True))
````

### Four Types of Module Import Styles in Python

| Import Style                  | Description                                                | Example                                  |
|:------------------------------|:-----------------------------------------------------------|:-----------------------------------------|
| `import module_name`          | Imports the entire module. Access members using `module_name.member`. | `import math` <br> `print(math.pi)`      |
| `import module_name as alias` | Imports the entire module and assigns it an alias for shorter reference. | `import pandas as pd` <br> `df = pd.DataFrame()` |
| `from module_name import object_name` | Imports only specific objects (functions, classes, variables) directly from the module. Access members without the module name. | `from math import pi` <br> `print(pi)`   |
| `from module_name import *`   | Imports all public objects from the module directly. Generally discouraged due to potential name clashes. | `from math import *` <br> `print(pi)`   |

### Differences among Function, Module, and Package

**Function:** A block of organized, reusable code that performs a single, related action. Defined using `def`, can take arguments, return values, and exists independently. For example, `len('hello')`, `print('world')`, or a custom function like `def add(a, b): return a + b`.

**Module:** A file (`.py` file) containing Python code (functions, classes, variables, etc.). Used for logically organizing code and reusing it by importing into other scripts. Examples include the `math` module (`math.pi`, `math.sqrt()`), the `datetime` module, or the `pandas` module.

**Package:** A way of organizing related modules into a directory hierarchy. Essentially a folder containing multiple modules, identified by an `__init__.py` file. It provides structured management for larger applications and prevents name clashes. You can import modules/sub-packages using dot notation, such as the `numpy` package, `scipy` package, or a custom structure like `my_project/src/utilities/__init__.py`, `my_project/src/utilities/data_processing.py`.

### Package Structure

A Python package is essentially a directory containing one or more Python modules and a special `__init__.py` file. The `__init__.py` file signals to Python that the directory should be treated as a package. Packages can also contain sub-packages, creating a hierarchical structure.

Here's a typical structure:

```
my_package/
├── __init__.py
├── module_a.py
├── sub_package_b/
│   ├── __init__.py
│   └── module_b1.py
└── sub_package_c/
    ├── __init__.py
    └── module_c1.py
```

### Advantages of Modules and Packages

Using modules and packages in Python offers several significant benefits for organizing and managing code:

1.  **Modularity and Organization**: They break down large programs into smaller, manageable, and logically organized files (modules) and directories (packages). This makes the code easier to understand, navigate, and maintain.

2.  **Code Reusability**: Functions, classes, and variables defined in a module can be reused across different projects or parts of the same project. This avoids code duplication and promotes efficiency.

3.  **Namespace Management**: Modules and packages help avoid naming conflicts (name collisions). Each module has its own namespace, meaning that names defined in one module won't clash with identical names in another module, provided they are accessed correctly (e.g., `math.pi` vs. `my_module.pi`).

4.  **Maintainability**: Changes made to a module or package are isolated and less likely to affect other parts of the codebase, as long as the interface remains consistent. This simplifies debugging and updates.

5.  **Collaboration**: In team environments, modules and packages allow multiple developers to work on different parts of a project simultaneously without interfering with each other's code.

6.  **Readability**: Well-structured code using modules and packages improves readability. It's easier to see the overall architecture of a project and locate specific functionalities.

7.  **Distribution**: Packages are the standard way to distribute Python libraries and applications to others. Tools like `pip` rely on this structure to install and manage external dependencies.

8.  **Hierarchy and Structure**: Packages introduce a hierarchical structure, allowing for deeper levels of organization (sub-packages). This is crucial for very large projects with many related components.

# 📂 Python File Handling – Quick Notes

## 🔹 1. Opening a File

```python
file = open("filename.txt", "mode")
```

### 📌 Modes

| Mode | Meaning           |
| ---- | ----------------- |
| 'r'  | Read (default)    |
| 'w'  | Write (overwrite) |
| 'a'  | Append            |
| 'x'  | Create new file   |
| 'b'  | Binary mode       |
| 't'  | Text mode         |

---

## 🔹 2. Closing a File

```python

```

## 🔹 3. Using `with` (Best Practice)

