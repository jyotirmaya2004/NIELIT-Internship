##Set

##### A set in python is an unordereas collection of unique elementd,meaning it does not allow duplicate values.
##### Since sets are unordered, they do not supports indexing or slicing.
##### However,they are mutable, allowing elements to be added or removed as needed.
##### Sets are defined using curly braces and are commonly used for performing mathematical set operations and ensuring data uniqueness.

```python
set1={1,2,4,2,4,2,53,5}
print(set1)
```

####Set modification

| Method      | Purpose                                                           | Example                            | Output / Effect            |
| ----------- | ----------------------------------------------------------------- | ---------------------------------- | -------------------------- |
| `add()`     | Adds **one element** to the set                                   | `s = {1, 2}`<br>`s.add(3)`         | `{1, 2, 3}`                |
| `update()`  | Adds **multiple elements** from iterable (list, tuple, set, etc.) | `s = {1, 2}`<br>`s.update([3, 4])` | `{1, 2, 3, 4}`             |
| `remove()`  | Removes a specific element. Gives **error if not found**          | `s = {1, 2, 3}`<br>`s.remove(2)`   | `{1, 3}`                   |
| `discard()` | Removes a specific element. **No error if not found**             | `s = {1, 2, 3}`<br>`s.discard(4)`  | `{1, 2, 3}`                |
| `pop()`     | Removes and returns a **random element**                          | `s = {1, 2, 3}`<br>`s.pop()`       | May remove `1` / `2` / `3` |
| `clear()`   | Removes **all elements** from the set                             | `s = {1, 2, 3}`<br>`s.clear()`     | `set()`                    |
| `copy()`   | Returns a **copy of set**                              | `s = {1, 2, 3}`<br>`s.copy()`     | `set(1,2,3)`                    |


```python
s={2,42,4,565,87,46}
s.add(45)
print(s)
```

```python
s.update([7,49])
print(s)
```

```python
s.remove(7)
```

```python
print(s.discard(10))
```

```python
s.pop()
print(s)
```

```python
s.clear()
```

```python
s1={1,2,4,5}
new_s=s1.copy()
print(s1)
print(new_s)
```

#### Set operations methods

| Method                   | Purpose                                                     | Example                                        | Result        |
| ------------------------ | ----------------------------------------------------------- | ---------------------------------------------- | ------------- |
| `union()`                | Combines all unique elements from both sets                 | `a = {1,2,3}`<br>`b = {3,4,5}`<br>`a.union(b)` | `{1,2,3,4,5}` |
| `intersection()`         | Returns only common elements                                | `a.intersection(b)`                            | `{3}`         |
| `difference()`           | Returns elements present in first set but not in second     | `a.difference(b)`                              | `{1,2}`       |
| `symmetric_difference()` | Returns elements that are in either set but **not in both** | `a.symmetric_difference(b)`                    | `{1,2,4,5}`   |


```python
a={2,53,56,35,22,75,98}
b={53,45,43,22,98,34}

print(a.union(b))
print(a.intersection(b))
print(a.difference(b))
print(a.symmetric_difference(b))
```

#### Set relation methods

| Method         | Purpose                                                      | Example                                         | Result |
| -------------- | ------------------------------------------------------------ | ----------------------------------------------- | ------ |
| `issubset()`   | Checks if all elements of one set are present in another set | `a = {1,2}`<br>`b = {1,2,3}`<br>`a.issubset(b)` | `True` |
| `issuperset()` | Checks if a set contains all elements of another set         | `b.issuperset(a)`                               | `True` |
| `isdisjoint()` | Checks if two sets have **no common elements**               | `a = {1,2}`<br>`c = {3,4}`<br>`a.isdisjoint(c)` | `True` |


```python
a={3,5,3,563,63}
b={3,63}
c={99,88}

print(a.issubset(b))
print(b.issubset(a))
print(a.issuperset(b))
print(a.isdisjoint(c))
```

##Introduction to functions in python

##### A function is a block of reusable code that performs a specific task and can be executed whenever it is called.

`def function_name(parameters):`
     <br>`#block of code`
      <br>`returns value`

####Types of functions

| Type of Function          | Meaning                                    | Syntax / Example                    | Output        |
| ------------------------- | ------------------------------------------ | ----------------------------------- | ------------- |
| **Built-in Function**     | Functions already provided by Python       | `print("Hello")`<br>`len("Python")` | `Hello`, `6`  |
| **User-defined Function** | Functions created by the user using `def`  | `def add():`                        | Custom output |
| **Lambda Function**       | Anonymous one-line function using `lambda` | `lambda x: x + 2`                   | Returns value |


#### Types of Arguments

| Type of Argument                                  | Meaning                                       | Example                    | Output             |
| ------------------------------------------------- | --------------------------------------------- | -------------------------- | ------------------ |
| **Positional Argument**                           | Values are passed according to position/order | `add(10, 20)`              | `30`               |
| **Keyword Argument**                              | Values are passed using parameter names       | `add(a=10, b=20)`          | `30`               |
| **Default Argument**                              | Uses a default value if no argument is passed | `greet()`                  | Default value used |
| **Variable-length Argument** (`*args`)            | Accepts multiple positional arguments         | `total(1,2,3)`             | `6`                |
| **Keyword Variable-length Argument** (`**kwargs`) | Accepts multiple keyword arguments            | `info(name="Ram", age=20)` | Dictionary output  |


```python
#1. Positional argument
def add(a,b):
  print("Positional : ",a+b)
add(7,8)
```

```python
#2. Keyword Argument
def add_keyword(a,b):
  print("Keyword : ",a+b)
  return a+b
add_keyword(b=7,a=8)
```

```python
#3. Default Argument
def greet(name="Guest"):
  print("Hello ",name)
greet()
greet("Jyotirmaya")
```

```python
#4. Variable length Argument
def total(*nums):
  print("Arguments : ",nums)
total(34,53,6,35,74,46)
```

```python
#5. Keyword variable length (**kwargs)
def info(**data):
  print("Keyword Arguments: ",data)
info(name="Jyotirmaya",Age=21)
```
