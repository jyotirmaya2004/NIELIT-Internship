##Dictionary

#### Access dictionary elements

| Method / Syntax                  | Description                                           | Example                 | Output                                              |
| -------------------------------- | ----------------------------------------------------- | ----------------------- | --------------------------------------------------- |
| `dict[key]`                      | Access value using key directly                       | `d["name"]`             | `"Jyotirmaya"`                                      |
| `dict.get(key)`                  | Access value safely (returns `None` if key not found) | `d.get("name")`         | `"Jyotirmaya"`                                      |
| `dict.get(key, default)`         | Returns default value if key not found                | `d.get("age", 0)`       | `0`                                                 |
| `dict.keys()`                    | Access all keys                                       | `d.keys()`              | `dict_keys(['name', 'age'])`                        |
| `dict.values()`                  | Access all values                                     | `d.values()`            | `dict_values(['Jyotirmaya', 20])`                   |
| `dict.items()`                   | Access key-value pairs                                | `d.items()`             | `dict_items([('name', 'Jyotirmaya'), ('age', 20)])` |
| `for key in dict`                | Access keys using loop                                | `for i in d:`           | `name`, `age`                                       |
| `for key in dict.keys()`         | Access keys explicitly                                | `for i in d.keys():`    | `name`, `age`                                       |
| `for value in dict.values()`     | Access only values                                    | `for i in d.values():`  | `Jyotirmaya`, `20`                                  |
| `for key, value in dict.items()` | Access both key and value                             | `for k,v in d.items():` | `name Jyotirmaya`                                   |
| Nested access                    | Access nested dictionary values                       | `d["student"]["name"]`  | `"Jyoti"`                                           |


```python
d={"name":"Jyotirmaya","age":20,"roll":25}
print(d["name"])

print(d.get("age"))
print(d.get("gender"))

print(d.get("gender","male"))

print(d.keys())
print(d.values())
print(d.items())

for i in d:
  print(i)

for i in d.keys():
  print(i)

for i in d.values():
  print(i)

for k,v in d.items():
  print(k,v)
#nested access
d={"student":{"name":"Jyoti","age":20}}
print(d["student"]["name"])
```

#### update dictionary elements

| Method / Syntax     | Description                | Example                                    | Result               |
| ------------------- | -------------------------- | ------------------------------------------ | -------------------- |
| `dict[key] = value` | Update existing key value  | `d["age"] = 21`                            | `{"age": 21}`        |
| `dict[key] = value` | Add new key-value pair     | `d["city"] = "Ranchi"`                     | `{"city": "Ranchi"}` |
| `dict.update()`     | Update one or more items   | `d.update({"age": 22})`                    | `{"age": 22}`        |
| `dict.update()`     | Add multiple items         | `d.update({"city":"Ranchi","pin":834001})` | Adds both            |
| `dict.setdefault()` | Add key only if not exists | `d.setdefault("country","India")`          | Adds only if missing |


```python
d={"name":"Jyotirmaya","age":20,"roll":25}
d["name"]="Jyotirmaya Behera"

print(d["name"])

d["city"]="Balasore"

print(d)

d.update({"age":21})
print(d)

d.update({"city":"Baleshwar","pin":756023})
print(d)

d.setdefault("country","India")
print(d)

user_profile = {"name": "Alice", "age": 28, "role":
"User"}
new_data     = {"age": 29, "location": "New York"}
user_profile.update(new_data)
print("Updated Profile:")
print(user_profile)
```

####Delete dictionary elements

| Method / Syntax          | Description                                      | Example                       | Result             |
| ------------------------ | ------------------------------------------------ | ----------------------------- | ------------------ |
| `del dict[key]`          | Delete a specific key-value pair                 | `del d["age"]`                | Removes `age`      |
| `dict.pop(key)`          | Removes key and returns its value                | `d.pop("age")`                | Removes `age`      |
| `dict.pop(key, default)` | Removes key safely, returns default if not found | `d.pop("phone", "Not Found")` | `"Not Found"`      |
| `dict.popitem()`         | Removes the last inserted item                   | `d.popitem()`                 | Removes last pair  |
| `dict.clear()`           | Removes all items from dictionary                | `d.clear()`                   | `{}`               |
| `del dict`               | Deletes entire dictionary                        | `del d`                       | Dictionary removed |


```python
inventory = {"apples": 10, "bananas": 5, "oranges": 8}
sold = inventory.pop("bananas")
print(f"Sold: {sold}")
print(f"Remaining: {inventory}")
pears = inventory.pop("pears", "Not found")
print(f"Pears: {pears}")
last_item = inventory.popitem()
print(f"Last item sold: {last_item}")
inventory.clear()
print(f"Inventory after clearing: {inventory}")
del inventory
```
