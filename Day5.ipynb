##List

#### Craetion Method

| Method                    | Syntax                              | Description                                | Example Output                                     |
| ------------------------- | ----------------------------------- | ------------------------------------------ | -------------------------------------------------- |
| Using square brackets     | `list_name = [ ]`                   | Directly creates a list using brackets     | `a = [1, 2, 3]` → `[1, 2, 3]`                      |
| Using `list()`            | `list_name = list(iterable)`        | Converts iterable objects into a list      | `a = list((1,2,3))` → `[1, 2, 3]`                  |
| Empty list using `[]`     | `list_name = []`                    | Creates an empty list                      | `a = []` → `[]`                                    |
| Empty list using `list()` | `list_name = list()`                | Another way to create an empty list        | `a = list()` → `[]`                                |
| List comprehension        | `[expression for item in iterable]` | Creates list in one line with loop logic   | `a = [x*x for x in range(5)]` → `[0, 1, 4, 9, 16]` |
| Using `range()`           | `list(range(start, stop, step))`    | Creates sequence of numbers as list        | `a = list(range(1, 6))` → `[1, 2, 3, 4, 5]`        |
| Using string              | `list("string")`                    | Converts each character into list elements | `a = list("hello")` → `['h','e','l','l','o']`      |
| Using `*` operator        | `[value] * n`                       | Repeats same element multiple times        | `a = [5] * 4` → `[5, 5, 5, 5]`                     |
| Using loop + append       | `for` loop with `.append()`         | Adds elements one by one                   | `a=[]; for i in range(3): a.append(i)` → `[0,1,2]` |
| Nested list               | `[[ ]]`                             | Creates list inside list                   | `a = [[1,2],[3,4]]` → `[[1,2],[3,4]]`              |


```python
my_list=["hello","World","Python"]
print(my_list)
my_list2=list((1,2,3))
print(my_list2)
my_list3=[]
print(my_list3)
my_list4=list()
print(my_list4)
my_list5=[x*x for x in range(5)]
print(my_list5)
my_list6=list(range(1,6))
print(my_list6)
my_list7=list("hello")
print(my_list7)
my_list8=[5]*4
print(my_list8)
a=[]
for i in range(5):
  a.append(i)
print(a)
my_list9=[[1,2],[3,4]]
print(my_list9)
```

#### Access Method

| Method             | Syntax                      | Description                                            | Example      | Output        |
| ------------------ | --------------------------- | ------------------------------------------------------ | ------------ | ------------- |
| Positive Indexing  | `list[index]`               | Access elements from left to right, starting from `0`  | `a[1]`       | `20`          |
| Negative Indexing  | `list[-index]`              | Access elements from right to left, starting from `-1` | `a[-1]`      | `40`          |
| Slicing            | `list[start:stop:step]`     | Access multiple elements in a range                    | `a[1:3]`     | `[20, 30]`    |
| Using Loop         | `for x in list`             | Access elements one by one directly                    | `for x in a` | `10 20 30 40` |
| Using Index + Loop | `for i in range(len(list))` | Access index and value together                        | `a[i]`       | `10 20 30 40` |


```python
a=[1,2,3,2,3,2,3,1432,5,45,4,65,767,68]
print(a[3])
print(a[-1])
print(a[1:3])
for i in a:
  print(i)
for i in range(len(a)):
  print(a[i])
```

#### Add element method

| Method     | Syntax                      | Description                                | Example           | Output             |
| ---------- | --------------------------- | ------------------------------------------ | ----------------- | ------------------ |
| `append()` | `list.append(value)`        | Adds an element at the **end** of the list | `a.append(40)`    | `[10, 20, 30, 40]` |
| `insert()` | `list.insert(index, value)` | Adds an element at a **specific position** | `a.insert(1, 15)` | `[10, 15, 20, 30]` |


```python
my_list=[10,20,30]
my_list.append(40)
print(my_list)
my_list.insert(1,15)
print(my_list)
```

#### Update element method

| Method                   | Syntax                       | Description                               | Example            | Output             |
| ------------------------ | ---------------------------- | ----------------------------------------- | ------------------ | ------------------ |
| Using Positive Index     | `list[index] = value`        | Updates element using left-to-right index | `a[1] = 50`        | `[10, 50, 30, 40]` |
| Using Negative Index     | `list[-index] = value`       | Updates element from right side           | `a[-1] = 99`       | `[10, 20, 30, 99]` |
| Using Slicing            | `list[start:stop] = values`  | Updates multiple elements at once         | `a[1:3] = [50,60]` | `[10, 50, 60, 40]` |
| Using Loop               | `for` loop with index        | Updates elements one by one               | `a[i] = a[i]*2`    | `[20,40,60,80]`    |
| Using List Comprehension | `[expression for x in list]` | Creates updated list in one line          | `[x*2 for x in a]` | `[20,40,60,80]`    |


```python
a=[22,33,55,66,77,88,99,44]
a[1]=100
print(a)
a[-1]=9999
print(a)
a[1:3]=[200,300]
print(a)
for i in range(len(a)):
  a[i]=a[i]*2
print(a)
a=[x*2 for x in a]
print(a)
```

#### delete method

| Method            | Syntax               | Description                             | Example        | Output         |
| ----------------- | -------------------- | --------------------------------------- | -------------- | -------------- |
| `remove()`        | `list.remove(value)` | Deletes a **specific value**            | `a.remove(20)` | `[10, 30, 40]` |
| `pop(index)`      | `list.pop(index)`    | Deletes element at a **specific index** | `a.pop(1)`     | `[10, 30, 40]` |
| `pop()`           | `list.pop()`         | Deletes the **last element**            | `a.pop()`      | `[10, 20, 30]` |
| `del` at index    | `del list[index]`    | Deletes element at given index          | `del a[2]`     | `[10, 20, 40]` |
| `del` entire list | `del list`           | Deletes the complete list               | `del a`        | list deleted   |
| `clear()`         | `list.clear()`       | Removes all elements, keeps empty list  | `a.clear()`    | `[]`           |


```python
a=[10,20,30,40,50,60,70,80,90]
a.remove(20)
print(a)
a.pop(1)
print(a)
a.pop()
print(a)
del a[2]
print(a)
del a
a=[10,20,30,40,50,60,70,80,90]
a.clear()
print(a)
```

## Tuple

#### Creation of tuple

| Method               | Syntax                | Description                       | Example                 | Output             |
| -------------------- | --------------------- | --------------------------------- | ----------------------- | ------------------ |
| Using parentheses    | `t = ( )`             | Creates a tuple using brackets    | `t = (1, 2, 3)`         | `(1, 2, 3)`        |
| Without parentheses  | `t = 1, 2, 3`         | Parentheses are optional          | `t = 1, 2, 3`           | `(1, 2, 3)`        |
| Single element tuple | `t = (value,)`        | Must use comma for one element    | `t = (5,)`              | `(5,)`             |
| Using `tuple()`      | `t = tuple(iterable)` | Converts iterable into tuple      | `t = tuple([1,2,3])`    | `(1, 2, 3)`        |
| Empty tuple          | `t = ()` or `tuple()` | Creates an empty tuple            | `t = ()`                | `()`               |
| Nested tuple         | `t = (( ),( ))`       | Tuple inside another tuple        | `t = ((1,2),(3,4))`     | `((1, 2), (3, 4))` |
| Using `range()`      | `tuple(range())`      | Creates tuple from sequence       | `t = tuple(range(1,5))` | `(1, 2, 3, 4)`     |
| Tuple packing        | `t = a, b, c`         | Multiple values packed into tuple | `t = 10, 20, 30`        | `(10, 20, 30)`     |
| Tuple unpacking      | `a,b,c = tuple`       | Assigns tuple values to variables | `a,b,c=(10,20,30)`      | `a=10,b=20,c=30`   |


```python

my_tuple=(1,2,3)
print("using parenthesis : ",my_tuple)
my_tuple2=1,2,3
print("Using without parenthesis : ",my_tuple2)
my_tuple3=(5,)
print("Single Element : ",my_tuple3)
my_tuple4=tuple([1,2,3])
print("Using tuple() : ",my_tuple4)
my_tuple5=()
print("Empty tuple : ",my_tuple5)
my_tuple6=((1,2),(3,4))
print("Nested tuple : ",my_tuple6)
my_tuple7=tuple(range(1,5))
print("using range() : ",my_tuple7)
a,b,c=10,20,30
print("Packing : ",a,b,c)
a,b,c = (10,20,30)
print("Unpacking : ",a,b,c)
```

#### Tuple unpacking method

| Method                  | Syntax                   | Description                                          | Example      | Output        |
| ----------------------- | ------------------------ | ---------------------------------------------------- | ------------ | ------------- |
| Positive Indexing       | `t[index]`               | Access element from left to right starting from `0`  | `t[1]`       | `20`          |
| Negative Indexing       | `t[-index]`              | Access element from right to left starting from `-1` | `t[-1]`      | `40`          |
| Slicing                 | `t[start:stop:step]`     | Access multiple elements                             | `t[1:3]`     | `(20, 30)`    |
| Using Loop              | `for x in t`             | Access elements one by one                           | `for x in t` | `10 20 30 40` |
| Using `range()` + index | `for i in range(len(t))` | Access by index in loop                              | `t[i]`       | `10 20 30 40` |


```python
t=tuple((23,1,5,53,56,353,56))
print(t[5])

print(t[-1])
print(t[1:5:1])
for i in t:
  print(i)
for i in range(len(t)):
  print(t[i])
```

#### Tuple update and delete tuple

| Operation           | Syntax                                  | Description                                      | Example                    | Output        |
| ------------------- | --------------------------------------- | ------------------------------------------------ | -------------------------- | ------------- |
| Direct update       | **Not allowed**                         | Tuples are immutable, elements cannot be changed | `t[1]=50`                  | Error         |
| Update (workaround) | Convert to list → modify → convert back | Used to update tuple indirectly                  | `list(t)`                  | Updated tuple |
| Delete element      | **Not directly possible**               | Must use conversion workaround                   | convert to list and delete | New tuple     |
| Delete entire tuple | `del tuple_name`                        | Deletes complete tuple                           | `del t`                    | tuple deleted |


```python
t=tuple((23,1,5,53,56,353,56))
tlist=list(t)
tlist[1]=9999
t=tuple(tlist)
print(t)
del t
```

## Dictionary

#### Craetion of dictionary

| Method                | Syntax                   | Description                             | Example                      | Output             |
| --------------------- | ------------------------ | --------------------------------------- | ---------------------------- | ------------------ |
| Using `{}`            | `d = {key:value}`        | Creates dictionary using curly braces   | `d = {"a":1,"b":2}`          | `{'a': 1, 'b': 2}` |
| Empty dictionary      | `d = {}`                 | Creates an empty dictionary             | `d = {}`                     | `{}`               |
| Using `dict()`        | `d = dict()`             | Creates empty dictionary using function | `d = dict()`                 | `{}`               |
| Using key-value pairs | `dict(key=value)`        | Creates dictionary with named keys      | `dict(a=1,b=2)`              | `{'a':1,'b':2}`    |
| Using list of tuples  | `dict([(k,v)])`          | Converts tuple pairs into dictionary    | `dict([("a",1),("b",2)])`    | `{'a':1,'b':2}`    |
| Using `zip()`         | `dict(zip(keys,values))` | Combines keys and values lists          | `dict(zip(k,v))`             | dictionary         |
| Nested dictionary     | `d = {key:{}}`           | Dictionary inside dictionary            | `{"student":{"name":"Ram"}}` | nested dict        |


```python
d={"a":1,"b":2,"c":3}
print(d)

d1={}
print("Empty dict ",d1)
d2=dict()
print("Empty dict ",d2)

d4=dict(a=2,b=3,c=4)
print(d4)

d5=dict([("a",1),("b",2)])
print(d5)

d6=dict(zip(["a","b","c"],[1,2,3]))
print("using zip()",d6)

d7={"student":{"name":"Ram"}}
print("Nested dictionary : ",d7)

```
